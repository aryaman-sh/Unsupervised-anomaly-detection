```python
for scan in scans:
    # check if same number of scans
    t1_nii_path = '/media/aryaman/T7/datasets/MICCAI_BraTS_2019_Data_Training/LGG/' + scan + '/' + scan + '_t1.nii'
    mask_nii_path = '/media/aryaman/T7/datasets/MICCAI_BraTS_2019_Data_Training/LGG/' + scan + '/' + scan + '_seg.nii'
    arr1 = nibabel.load(t1_nii_path).get_data()
    arr2 = nibabel.load(mask_nii_path).get_data()
    if arr1.shape[2] != arr2.shape[2]:
        print(f'mismatch {scan}')
        continue
    
    os.makedirs('/home/aryaman/png_brats_LGG/'+scan)
    # Convert T1w images
    # Create T1w dir
    os.makedirs('/home/aryaman/png_brats_LGG/'+scan+'/'+'t1w')
    t1_nii_path = '/media/aryaman/T7/datasets/MICCAI_BraTS_2019_Data_Training/LGG/' + scan + '/' + scan + '_t1.nii'
    image_array = nibabel.load(t1_nii_path).get_data()
    nx, ny, nz = image_array.shape
    total_slices = image_array.shape[2]
    for current_slice in range(total_slices):
        image_name = str(current_slice+1)+".png"
        data = numpy.rot90(image_array[:,:,current_slice])
        imageio.imwrite(image_name, data)
        src = image_name
        shutil.move(src, '/home/aryaman/png_brats_LGG/'+scan+'/'+'t1w')
        
    # Convert masks
    os.makedirs('/home/aryaman/png_brats_LGG/'+scan+'/'+'mask')
    mask_nii_path = '/media/aryaman/T7/datasets/MICCAI_BraTS_2019_Data_Training/LGG/' + scan + '/' + scan + '_seg.nii'
    image_array = nibabel.load(mask_nii_path).get_data()
    nx, ny, nz = image_array.shape
    total_slices = image_array.shape[2]
    for current_slice in range(total_slices):
        image_name = str(current_slice+1)+".png"
        data = numpy.rot90(image_array[:,:,current_slice])
        imageio.imwrite(image_name, data)
        src = image_name
        shutil.move(src, '/home/aryaman/png_brats_LGG/'+scan+'/'+'mask')
```
